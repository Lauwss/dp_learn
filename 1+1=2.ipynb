```python

import torch
```

```python
def add(A,B):
    return A + B
def subtract(A,B):
    return A - B
def multiply(A,B):
    return A * B
def devide (A,B):
    return A / B
```

```python
from torch import nn

a=torch.tensor([1.0],requires_grad=True)
b=torch.tensor([1.0],requires_grad=True)
loss=nn.MSELoss()
```

```python
trainer=torch.optim.SGD([a,b], lr=0.08)
num_epochs=30000
batch_size=5
from tqdm import tqdm
for epoch in tqdm(range(num_epochs),desc='训练中',):
        l_a = loss(add(a,b), torch.tensor([90.0]))
        l_b=  loss(devide(a,b), torch.tensor([17.0]))
        l = l_a + l_b
        trainer.zero_grad()
        l.backward()
        trainer.step()
        if (epoch + 1) % 5000 == 0:
                print(f'epoch {epoch + 1}, loss {l.item():.6f}, a {a.item():.6f}, b {b.item():.6f}')
                

      
        
```
